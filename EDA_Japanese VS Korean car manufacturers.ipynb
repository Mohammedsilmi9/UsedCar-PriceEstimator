Abstract:

There are many aspects to consider about a particular car maker or model when it comes to buying a used vehicle. Cost of ownership, reliability and even resale value. Here, we are going to inspect 3 important element that are worth examining.

1) Price: Throughout the collected data available, we are able to do a straight forward price comparison to Toyota against Hyundai models by year of manufacturing. 

2) Reliability: While we have gathered mileage information about the vehicles listed. It is safe to say that the higher the average mileage is
on a particular model the higher it ranks on the reliability scale. 

3) Robustness: Each vehicle listed for sale has a condition and title states attribute that we derived a score from. the better the condition the higher the score will be. Also vehicles with clean titles rank higher than vehicles with salvaged titles.

```python
from dataWrangler import *
from IPython.display import Image
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import glob 
import os
```

```python
# loading raw csv files from local dirctory of all makes and models. 
li=[]
for filename in glob.glob(r"..\UsedCar-PriceEstimator\data\*.csv"):
    collection = pd.read_csv(filename, index_col=None, header=0)
    li.append(collection)
frame = pd.concat(li, axis=0, ignore_index=True)
```

### We will be comparing vehicles that belong to the same tier

#### hyundai accent VS toyota corolla

```python
#choosing two vehicles to compare 
vehicle1='hyundai accent'
vehicle2='toyota corolla'
#calling the compare function
result=compare_2vehicles(vehicle1,vehicle2,frame)
```

```python
# life time and reliablilty test
_=result.plot(kind='bar',x='Year',y=[result.columns[2],result.columns[-2]])
```

Early model corollas have gone to higher mileage numbers than accents,
however, hyundais have made a decent come back after 2008 to be more in use as they seem to gain more popularity due to their low prices.

```python
_=result.plot(kind='bar',x='Year',y=[result.columns[3],result.columns[-1]])
```

```python
result["toyota corolla price"].mean()/result["hyundai accent price"].mean()
```

Corolla's on average cost x1.6 more than Accents.

```python
#robustness test
_=result.plot(x='Year',y=[result.columns[1],result.columns[-3]]
            ,alpha=0.6,bins=30,kind='hist')
```

```python
result["hyundai accent score"].mean(),result["toyota corolla score"].mean()
```

Accents are spread accross the condition scale and corollas seem to hold mid average rates

#### hyundai elantra VS toyota camry

```python
#choosing two vehicles to compare 
vehicle1='hyundai elantra'
vehicle2='toyota camry'
#calling the compare function
result=compare_2vehicles(vehicle1,vehicle2,frame)
```

```python
# life time and reliablilty test
_=result.plot(kind='bar',x='Year',y=[result.columns[2],result.columns[-2]])
```

It seems like camry owners arent listing their cars for sale until
